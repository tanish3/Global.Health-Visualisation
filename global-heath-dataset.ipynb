```python
geolocator = Nominatim(user_agent="geoapiExercises")
```

```python
def location(lat, long):
    location = geolocator.reverse(str(lat)+","+str(long))
    if location[0].split(",")[-3] == ' New York':
        return True
    else:
        return False
```

```python
data_ = pd.DataFrame(columns = data.columns)
data_path = glob("../input/global-health-dataset/gh_2021-08-16/*")
data_path.remove('../input/global-health-dataset/gh_2021-08-16/data_dictionary.csv')
data_path.remove('../input/global-health-dataset/gh_2021-08-16/citation_data.rtf')
for x in data_path:
    df = pd.read_csv(x)
    print(x)
    for x in range(len(df)):
        if location(df.loc[x]['location.geometry.latitude'], df.loc[x]['location.geometry.longitude']):
            data_.append(df.loc[x], ignore_index=True)
            print(data_)
data_.to_csv('New_york_data.csv', index=False)
        
# data = pd.concat(datalist)
# print(datalist)
```
